#### (2)-
##### Scatter plots are used to observe the relationship between variables and uses dots to represent this relationship.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize= (6,3))
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Scatter plot between x and y")
```

#### (3)-
##### subplot() function is used to create multiple plots in a single figure side by side. The subplot() function takes 3 arguments that describes the layout of figures- the first and second argument describe the row and column, the third argument represents the index of the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

#For line 1: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)                  # 2 rows and 2 columns, the 1st plot is at index=1 i.e. 1st position
plt.plot(x,y)                       # it will generate a line plot
plt.title("line 1")

#For line 2: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)                  # 2 rows and 2 columns, the 2nd plot is at index=2 i.e. 2nd position
plt.plot(x,y)
plt.title("line 2")

#For line 3: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)                 # 2 rows and 2 columns, the 3rd plot is at index=3 i.e. 3rd position
plt.plot(x,y) 
plt.title("line 3")

#For line 4: 
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)                 # 2 rows and 2 columns, the 4th plot is at index=4 i.e. 4th position
plt.plot(x,y)
plt.title("line 4")
```

#### (4)-
##### A bar plot shows catergorical data as rectangular bars with heights proportional to the value they represent. It is often used to compare between values of different categories in the data. In order to plot categorical data vs. numerical data, bar plot is used.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(6,4))
plt.bar(company, profit)
plt.xlabel("Company (categorical data)")
plt.ylabel("Profit (numerical data)")
plt.title("Bar plot")
```

```python
plt.figure(figsize=(6,4))
plt.barh(company, profit)                   # this will display a horizontal bar graph
plt.xlabel("Profit (numerical data)")
plt.ylabel("Company (categorical data)")
plt.title("Horizontal Bar plot")
```

#### (5)-
##### A box plot which is also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum.
##### Box plots are used to show distributions of numeric data values, especially when we want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data’s symmetry, skew, variance, and outliers. It is easy to see where the main bulk of the data is, and make that comparison between different groups.


```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data= [box1, box2]

fig= plt.figure(figsize= (6,4))
ax= fig.add_axes([0,0,1,1])
bp= ax.boxplot(data)
plt.show()
```
